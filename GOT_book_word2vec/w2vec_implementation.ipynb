```python
import gensim
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story=[]
for filename in os.listdir("book"):
    #print(filename)
    with open('book/{}'.format(filename),encoding='utf-8',
                 errors='ignore') as f:
        #print(f)
        corpus=f.read()
        raw_sent=sent_tokenize(corpus)
        for sent in raw_sent:
            story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
story
```

```python
story[0]
```

```python
model=gensim.models.Word2Vec(
    window=10,
    min_count=2,
    workers=4,
)
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
# first pass the entire corpus
# second total examples in corpus by model.corpus_count
# third number of epochs 
```
